```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('dataset.csv')

# Custom color palette
custom_palette = sns.color_palette("husl", 3)

# Data Cleaning and Processing (if necessary)

# Summary Statistics
summary_stats = df.describe()

# Set up subplots with increased width and adjusted horizontal space
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 12), gridspec_kw={'hspace': 0.6, 'wspace': 0.3})

# Visualization 1: Distribution of Student Outcome (Pie Chart)
outcome_counts = df['Target'].value_counts()
explode = (0.1, 0, 0)  # Explode the first slice for emphasis
axes[0, 0].pie(outcome_counts, labels=outcome_counts.index, autopct='%1.1f%%', startangle=90, explode=explode, colors=custom_palette)
axes[0, 0].set_title('Distribution of Student Outcome')
axes[0, 0].text(0.5, -0.15,
    "• 2209 students successfully graduated.\n"
    "• 1421 dropped out.\n"
    "• 794 are currently enrolled.", size=10, ha="left", va="center", transform=axes[0, 0].transAxes)

# Visualization 2: Curricular Units 1st Semester Grade vs Student Outcome
sns.boxplot(x='Target', y='Curricular units 1st sem (grade)', data=df, ax=axes[0, 1], palette=custom_palette)
axes[0, 1].set_title('Curricular Units 1st Semester Grade vs Student Outcome')
axes[0, 1].set_xlabel('Outcome')
axes[0, 1].set_ylabel('Curricular Units 1st Semester Grade')
axes[0, 1].text(0.5, -0.40, "Graduate students have higher grades \n as compared to the dropout students.", size=10, ha="left", va="center", transform=axes[0, 1].transAxes)

# Rotate x-axis labels in the second column of subplots
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=45, ha='right')

# Visualization 3: Distribution of Age at Enrollment for Student Outcome
age_distribution_text = (
    "The histplot indicates:\n"
    "• A peak in enrollment at ages 18-19.\n"
    "• A consistent high count of graduates across various age groups.\n"
    "• Dropout counts vary, with higher numbers in late teens and early twenties.\n"
    "• Enrollment decreases with age.\n"

)
sns.histplot(data=df, x='Age at enrollment', hue='Target', multiple='stack', bins=20, ax=axes[1, 0], palette=custom_palette)
axes[1, 0].set_title('Distribution of Age at Enrollment for Student Outcome')
axes[1, 0].set_xlabel('Age at Enrollment')
axes[1, 0].set_ylabel('Count')
axes[1, 0].text(0.4, -0.27, age_distribution_text, size=10, ha="left", va="center", transform=axes[1, 0].transAxes)

# Visualization 4: Impact of Scholarship on Student Outcome
sns.countplot(x='Target', hue='Scholarship holder', data=df, ax=axes[1, 1], palette=custom_palette)
axes[1, 1].set_title('Impact of Scholarship on Student Outcome')
axes[1, 1].set_xlabel('Outcome')
axes[1, 1].set_ylabel('Count')
axes[1, 1].legend(title='Scholarship Holder', labels=['No', 'Yes'])
axes[1, 1].text(0.3, -0.29, "Students with scholarships, whether present or not tend to have higher \n graduation counts compared to enrollments and dropouts.", size=10, ha="left", va="center", transform=axes[1, 1].transAxes)

# Rotate x-axis labels in the second column of subplots
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45, ha='right')

# Add a  title
fig.suptitle("College Dropout and Success Analysis (Sadaf Naz, id: 22080502)", size=16)




# Save Infographic
plt.show()



```
